```python
# 0721, remove duplicates
import pandas as pd
import distance
```

```python
ieee_df = pd.read_csv('IEEE_total_20170801.csv')
springer_df = pd.read_csv('Springer_total_20170801.csv')
ACM_scopus_df = pd.read_csv('scopus_total_20170801.csv')
ACM_scopus_df.head()
```

```python
spinger_titles = springer_df['Item Title']
ieee_title = ieee_df['Document Title']
acm_title = ACM_scopus_df['Title']
print len(spinger_titles)
print len(ieee_title)
print len(acm_title)
```

```python
print set(ieee_title)
```

```python
def compare_string(set1,set2):
    duplicates1=set()
    duplicates2=set()
    for title1 in set1:
        title1_norm = title1.lower()
        title1_norm = title1_norm.strip()
        title1_norm = title1_norm.split(' ')
        
        for title2 in set2:
            title2_norm = title2.lower()
            title2_norm = title2_norm.strip()
            title2_norm = title2_norm.split(' ')
            if len(title2_norm)== len(title1_norm):
                if distance.levenshtein(title1_norm, title2_norm)<2:
                    print('title1',title1)

                    print('title2',title2)
                    print('\n')
                    duplicates1.add(title1)
                    duplicates2.add(title1)
    return duplicates1, duplicates2
                
    
springer_set = set(spinger_titles)
ieee_set = set(ieee_title)
acm_set = set(acm_title)
ieee_dup,acm_dup = compare_string(ieee_set,acm_set)       
```

```python
for title in ieee_dup:
    ieee_df=ieee_df[ieee_df['Document Title']!=title]

len(ieee_df)
```

```python
springer_dup,acm_dup = compare_string(springer_set,acm_set)    
```

```python
for title in springer_dup:
    springer_df=springer_df[springer_df['Item Title']!=title]

len(springer_df)
```

```python
springer_dup,ieee_dup = compare_string(springer_set,ieee_set)  
```

```python
for title in springer_dup:
    springer_df=springer_df[springer_df['Item Title']!=title]

len(springer_df)
```

```python
springer_df.to_csv('Springer_total_20170821_cleaned.csv')
ieee_df.to_csv('IEEE_total_20170821_cleaned.csv')
```
