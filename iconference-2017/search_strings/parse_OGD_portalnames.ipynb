```python
def clean_html(url):
    if 'http://' in url:
        url =url.replace("http://","")
    if 'https://' in url:
        url =url.replace("https://","")
    if 'www.' in url:
        url =url.replace("www.","")
    search_term = url[0:-1]
    

    double_q = '"' # double quote
    search_term = double_q + search_term + double_q
        
        
    return search_term
```

```python
link_list = list(port_df['Link'].map(clean_html))
link_list
```
